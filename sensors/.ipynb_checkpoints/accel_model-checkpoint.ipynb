```python
from MPU6050 import MPU6050
import numpy as np

timeVector = np.arange(0,100,0.1)
accel = MPU6050(time=timeVector, time_step=0.1, duration=100, mode="accelerometer_only", 
low_power_wakeup=0, loop_rate=60, digital_low_pass=0, sample_rate_divisor=255)
```

```python
times = [(0,5.6,"accelerometer_only"), (5.6,25,"gyroscope_accelerometer_DMP"), (25,100,"low_power_wakeup_1.25")]
power, data = accel.getPowerVector(times)
accel.plotData(power, data, timeVector)
```
